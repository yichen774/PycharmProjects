```python
import cv2
import numpy as np
img = cv2.imread('lewis-hamilton-mercedes-1.jpg')
img = cv2.resize(img, (0,0), fx = 0.7, fy = 0.7)
#創造一個都是1的二維函數
kernel = np.ones((3,3), np.uint8)

#灰階的圖片
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#高斯模糊 (img,(基數,基數), 標準差)
blur = cv2.GaussianBlur(img, (11,11), 4)
#描邊(邊緣最低門檻,最高門檻)
canny = cv2.Canny(img, 150,200)
#膨脹把線條變粗
dilate = cv2.dilate(canny, kernel, iterations = 2)

erode = cv2.erode(dilate, kernel,iterations = 2 )


cv2.imshow('img',img)
cv2.imshow('gray',gray)
cv2.imshow('blur',blur)
cv2.imshow('canny',canny)
cv2.imshow('dilate',dilate)
cv2.imshow('erode',erode)
cv2.waitKey(0)
```

```python
import cv2
import numpy as np

img = cv2.imread("lewis-hamilton-mercedes-1.jpg")

def empty(v):
    pass


cv2.namedWindow('TrackBar')
cv2.resizeWindow("TrackBar",600,300)

cv2.createTrackbar("Hue Min","TrackBar",0,179,empty)
cv2.createTrackbar("Hue Max","TrackBar",0,179,empty)
cv2.createTrackbar("Sat Min","TrackBar",0,179,empty)
cv2.createTrackbar("Sat Max","TrackBar",0,179,empty)
cv2.createTrackbar("Val Min","TrackBar",0,179,empty)
cv2.createTrackbar("Val Max","TrackBar",0,179,empty)
#("控制條名稱","哪個視窗底下","初始值",""最大值,"控制條被改變後所帶的函式")

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)#轉換顏色

while True:
    h_min = cv2.getTrackbarPos('Hue Min', 'TrackBar')
    h_max = cv2.getTrackbarPos('Hue Max', 'TrackBar')
    s_min = cv2.getTrackbarPos('Sat Min', 'TrackBar')
    s_max = cv2.getTrackbarPos('Sat Max', 'TrackBar')
    v_min = cv2.getTrackbarPos('Val Min', 'TrackBar')
    v_max = cv2.getTrackbarPos('Val Max', 'TrackBar')
    print(h_min,h_max,s_min,s_max,v_min,v_max)
    
    lower = np.array([h_min,s_min,v_min])
    upper = np.array([h_max,s_max,v_max])
    
    mask = cv2.inRange(hsv,lower,upper)#(過濾的圖片,最小值,最大值)
    result = cv2.bitwise_and(img,img, mask=mask)#把兩張圖片的bit相加再套上mask作為遮罩
    
    cv2.imshow("mask",mask)
    cv2.imshow("img", img)
    cv2.imshow("hsv", hsv)
    cv2.imshow("result",result)
    cv2.waitKey(1)
    
    

```

```python
import cv2

img = cv2.imread("anyshape.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)#轉換灰階
imgContour = img.copy() #複製一張圖片讓輪廓點畫上去
canny = cv2.Canny(img,250,260)
contours,hierarchy = cv2.findContours(canny ,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
#偵測輪廓(偵測圖片,偵測模式，近似方法)

for cnt in contours:
    cv2.drawContours(imgContour,cnt,-1,(2,33,33),4)
    #劃出輪廓點在複製的圖片上(被畫的圖片，點的位置,要畫得輪廓是第幾個(-1=全畫),RGB,粗度)
    area = cv2.contourArea(cnt)
    if area > 500:
        peri = cv2.arcLength(cnt, True)#輪廓邊長(輪廓點,是否閉合)
        vertices = cv2.approxPolyDP(cnt, peri*0.02 ,True)
        #多邊形近似輪廓 近似值越大多邊形越大
        corners = len(vertices)
        x,y,z,h = cv2.boundingRect(vertices)
        cv2.rectangle(imgContour,(x,y),(x+z,y+h),(0,255,0),4)
        if corners == 3:
            cv2.putText(imgContour,"triangle",(x,y-5),cv2.FONT_ITALIC,1,(0,255,0),2)
        elif corners ==4:
            cv2.putText(imgContour,"rectangle",(x,y-5),cv2.FONT_ITALIC,1,(0,255,0),2)
        else:
            cv2.putText(imgContour,"Ed Shareen",(x,y-5),cv2.FONT_ITALIC,1,(0,255,0),2)
# cv2.imshow("img",img)
# cv2.imshow("canny",canny)
cv2.imshow("Contour",imgContour)
cv2.waitKey(0)
```

```python
import cv2

img = cv2.imread('f12022_drivers.jpg')
img = cv2.resize(img, (0,0), fx = 0.8, fy=0.8)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faceCascade = cv2.CascadeClassifier('face_detect.xml')
faceRect = faceCascade.detectMultiScale(gray,1.001,7)
print(len(faceRect))

for (x,y,w,h)in faceRect:
    cv2.rectangle(img, (x,y),(x+w,y+h),(0,255,0),2)

cv2.imshow("img",img)
cv2.waitKey(0)
```
